<h3 style="color:#0071DC; text-align:center; background-color:#e7e9eb;">** Data preparation ** </h3> 

```python
# Dropping unnecessary columns

coaster_df = coaster_df[['coaster_name',
    # 'Length', 'Speed',
    'Location', 'Status',
    # 'Opening date',
    #   'Type',
    'Manufacturer',
#     'Height restriction', 'Model', 'Height','Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
#      'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle', 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
#      'Track layout', 'Fastrack available', 'Soft opening date.1',
#      'Closing date', 'Opened', 'Replaced by', 'Website', 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
#      'Single rider line available', 'Restraint Style', 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',        
       'year_introduced', 'latitude', 'longitude', 'Type_Main', 'opening_date_clean',       
#       'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
#       'height_value', 'height_unit',
        'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()

coaster_df.shape
```

```python
# Changing 'opening_date_clean' column data type to datetime

coaster_df['opening_date_clean'] = pd.to_datetime(coaster_df['opening_date_clean'])
coaster_df.dtypes
```

```python
# Rename columns
coaster_df = coaster_df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})
```

```python
# Count the number of missing values
coaster_df.isna().sum()
```

```python
# Checking and locating duplicates

coaster_df.loc[coaster_df.duplicated()]
```

```python
# Checking an example duplicate
coaster_df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# Remove duplicates based on 'Coaster_Name','Location' and 'Opening_Date' columns
newdf = coaster_df.drop_duplicates( 
  subset = ['Coaster_Name','Location','Opening_Date'], 
  keep = 'last').reset_index(drop = True)

newdf.duplicated().sum()
```

<h3 style="color:#0071DC; text-align:center; background-color:#e7e9eb;">** Questions to Explore ** </h3>

<h4 style="color:#0071DC; background-color:#e7e9eb;">What are the top 10 fastest roller coasters, and how do they compare in terms of height?</h4>

```python
# Sort the dataset by speed and select the top 10 fastest coasters
top_fastest_coasters = newdf.sort_values(by='Speed_mph', ascending=False).head(10)

# Plotting speed vs height for the top 10 fastest coasters
plt.figure(figsize=(10, 6))
sns.barplot(x='Speed_mph', y='Coaster_Name', data=top_fastest_coasters, palette='coolwarm')
plt.title('Top 10 Fastest Roller Coasters')
plt.xlabel('Speed (mph)')
plt.ylabel('Coaster Name')
plt.show()

# Compare their heights
plt.figure(figsize=(10, 6))
sns.barplot(x='Height_ft', y='Coaster_Name', data=top_fastest_coasters, palette='coolwarm', order=top_fastest_coasters.sort_values('Height_ft',ascending = False).Coaster_Name)
plt.title('Height of the Top 10 Fastest Roller Coasters')
plt.xlabel('Height (ft)')
plt.ylabel('Coaster Name')
plt.show()
```

<h4 style="color:#0071DC; background-color:#e7e9eb;">Is there a notable trend in roller coaster speed or height over the years?</h4>

```python
group_by_year = newdf.groupby('Year_Introduced')['Speed_mph','Height_ft'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.lineplot(group_by_year, x='Year_Introduced', y='Speed_mph', marker='o')

plt.figure(figsize=(8, 6))
sns.lineplot(group_by_year, x='Year_Introduced', y='Height_ft', marker='o')
```

<h4 style="color:#0071DC; background-color:#e7e9eb;">What is the distribution of roller coasters' speeds and heights?</h4>

```python
# Plotting the distrubution of coaster's speed 

plt.figure(figsize=(8, 6))

sns.histplot(newdf['Speed_mph'], kde=True)
plt.title('Coaster Speed Distribution')
plt.show()

# Plotting the distribution of coaster's height
sns.histplot(newdf['Height_ft'], kde=True)
plt.title('Coaster Height Distribution')
plt.show()
```

<h4 style="color:#0071DC; background-color:#e7e9eb;">Which roller coaster manufacturers have the highest average speed or height?</h4>

```python
# Speed and height trend by manufacturer

group_by_manufacturer = newdf.groupby('Manufacturer')['Speed_mph','Height_ft'].mean().reset_index().sort_values(by='Speed_mph', ascending=False).head(10)

# Plotting speed vs height for the top 10 manufacturer
plt.figure(figsize=(10, 6))
sns.barplot(x='Speed_mph', y='Manufacturer', data=group_by_manufacturer, palette='coolwarm')
plt.title('Top 10 Fastest Manufacturer')
plt.xlabel('Speed (mph)')
plt.ylabel('Manufacturer Name')
plt.show()

# Compare their heights
plt.figure(figsize=(10, 6))
sns.barplot(x='Height_ft', y='Manufacturer', data=top_fastest_coasters, palette='coolwarm')
plt.title('Height of the Top 10 Fastest Manufacturer')
plt.xlabel('Height (ft)')
plt.ylabel('Manufacturer Name')
plt.show()
```
